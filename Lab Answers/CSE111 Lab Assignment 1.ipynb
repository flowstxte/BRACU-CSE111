```python
#Task 1*
s=input()
uc=0
lc=0
for i in s:
  if 65<=ord(i)<=90:
    uc+=1
  elif 97<=ord(i)<=122:
    lc+=1
if uc<=lc:
  for i in s:
    if 65<=ord(i)<=90:
      print(chr(ord(i)+32),end='')
    else:
      print(i,end='')
else:
  for i in s:
    if 97<=ord(i)<=122:
      print(chr(ord(i)-32),end='')
    else:
      print(i,end='')
```

```python
#Task 2*
s=input()
ch=False
nm=False
for i in s:
  if 65<=ord(i)<=90 or 97<=ord(i)<=122:
    ch=True
  elif 48<=ord(i)<=57:
    nm=True
if ch==True and nm==True:
  print('MIXED')
elif nm==True:
  print('NUMBER')
elif ch==True:
  print('WORD')
```

```python
#Task 5*
s=input()
sp='_$#@'
low=False
up=False
num=False
spc=False
for i in s:
  if 97<=ord(i)<=122:
    low=True
  elif 65<=ord(i)<=90:
    up=True
  elif 48<=ord(i)<=57:
    num=True
  elif i in sp:
    spc=True
x=''
if low==True and up==True and num==True and spc==True:
  x+='OK, '
if low==False:
  x+='Lowercase character missing, '
if up==False:
  x+='Uppercase character missing, '
if num==False:
  x+='Digit missing, '
if spc==False:
  x+='Special character missing, '
print(x[:-2])
```

#List

```python
#Task 1*
l1=[]
l2=[]
while True:
  x=input()
  if x!='STOP':
    l1.append(int(x))
  else:
    break
for i in l1:
  if i not in l2:
    l2.append(i)
for i in l2:
  c=0
  for j in l1:
    if i==j:
      c+=1
  print(i,'-',c,'times')
```

```python
#Task 2*
l=[]
for i in range(int(input())):
  l.append([int(i) for i in input().split(' ')])
max=0
for i in l[0]:
  max+=i
for i in range(len(l)):
  max2=0
  for j in l[i]:
    max2+=j
  if max2>max:
    max=max2
    x=i
print(max)
print(l[x])
```

```python
#Task 4*
while True:
  x=input()
  if x!='STOP':
    l=[int(i) for i in x.split(' ')]
    for i in range(len(l)-1):
      if l[i]>l[i+1]:
        if (l[i]-l[i+1])<=(len(l)-1):
          v=True
        else:
          v=False
      elif l[i+1]>=l[i]:
        if (l[i+1]-l[i])<=(len(l)-1):
          v=True
        else:
          v=False
      else:
        v=False
    if v==True:
      print('UB Jumper')
    else:
      print('Not UB Jumper')
  else:
    break
```

#Dictionary & Tuple

```python
#Task 1*
d1={}
d2={}
s1=input().split(', ')
s2=input().split(', ')
for i in s1:
  x1=i.split(': ')
  d1[x1[0]]=int(x1[1])
for i in s2:
  x2=i.split(': ')
  d2[x2[0]]=int(x2[1])
dn={}
for i,j in d1.items():
  if i not in dn:
    dn[i]=j
  else:
    dn[i]+=j
for i,j in d2.items():
  if i not in dn:
    dn[i]=j
  else:
    dn[i]+=j
print(dn)
x=[]
for i in dn.values():
  if i not in x:
    x.append(i)
print('Values:',tuple(sorted(x)))
```

```python
#Task 2*
d={}
while True:
  x=input()
  if x!='STOP':
    if int(x) not in d:
      d[int(x)]=1
    else:
      d[int(x)]+=1
  else:
    break
for i,j in d.items():
  print(i,'-',j,'times')
```

```python
#Task 3*
d1={}
s=input().split(', ')
for i in s:
  x=i.split(' : ')
  d1[x[0]]=x[1]
d2={}
for i,j in d1.items():
  if j not in d2:
    d2[j]=[i]
  else:
    d2[j].append(i)
print(d2)
```

```python
#Task 4*
n=input()
s=''
for i in n:
  if 97<=ord(i)<=122:
    s+=chr(ord(i)-32)
  else:
    s+=i
x=''
d={'1':'.,?!:','2':'ABC','3':'DEF','4':'GHI','5':'JKL',
   '6':'MNO','7':'PQRS','8':'TUV','9':'WXYZ','0':' '}
for i in s:
  for key,val in d.items():
    if i in val:
      for j in val:
        x+=key
        if i==j:
          break
print(x)
```
