```python
#Task 1
def divider(t):
  l=[int(i) for i in t.split(', ')]
  fn=l[0]
  ln=l[1]
  if ln==0:
    return 0
  elif fn==0:
    return 0
  else:
    return (fn/ln)-(fn//ln)
print(divider(input()))
```

```python
#Task 1*
def divider(n1, n2):
  if n1==0:
    return 0
  elif n2==0:
    return 0
  else:
    return (n1/n2)-(n1//n2)
print(divider(5, 2))
```

```python
#Task 3
def divsum(t):
  l=[int(i) for i in t.split(', ')]
  fn=l[0]
  ln=l[1]
  div=l[2]
  sum=0
  for i in range(fn,ln):
    if i%div==0:
      sum+=i
  return sum
print(divsum(input()))
```

```python
#Task 3*
def divsum(n1, n2, n3):
  sum=0
  for i in range(n1,n2):
    if i%n3==0:
      sum+=i
  return sum
print(divsum(0, 10, 2))
```

```python
#Task 5
def replace_domain(x,old='kaaj.com'):
  l=x.split(', ')
  if len(l)==2:
    mail=l[0]
    new=l[1]
  else:
    mail=l[0]
    new=l[1]
    old=l[2]
  y=mail.split('@')
  if y[1]==old:
    return f"Changed: {y[0]+'@'+new}"
  else:
    return f"Unchanged: {mail}"
print(replace_domain(input()))
```

```python
#Task 6
def vowel_count(s):
  vow='AEIOUaeiou'
  c=0
  x=''
  for i in s:
    if i in vow:
      c+=1
      x+=i+', '
  if c==0:
    print('No vowels in the name')
  else:
    print(f'Vowels: {x[:-2]}. Total number of vowels: {c}')
vowel_count(input())
```

```python
#Task 7
def palindrome(s):
  x=''
  for i in s:
    if i!=' ':
      x+=i
  y=''
  for i in range(1,len(x)+1):
    y+=x[-i]
  if x==y:
    return 'Palindrome'
  else:
    return 'Not a palindrome'
print(palindrome(input()))
```

```python
#Task 9
def capital(s):
  x=''
  ch='.!?'
  for i in range(len(s)):
    if i==0:
      x+=chr(ord(s[i])-32)
    elif i>=2 and s[i-2] in ch:
      x+=chr(ord(s[i])-32)
    elif s[i]=='i' and s[i+1]==' ' and s[i-1]==' ':
      x+=chr(ord(s[i])-32)
    else:
      x+=s[i]
  return x
print(capital(input()))
```
