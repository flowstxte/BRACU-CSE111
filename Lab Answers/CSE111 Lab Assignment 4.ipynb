```python
#Task 14
class StudentDatabase:
  def __init__(self,name,id):
    self.name=name
    self.id=id
    self.grades={}
  def calculateGPA(self,cl,sem):
    l=[]
    s=0
    d={}
    for i in cl:
      x=i.split(': ')
      l.append(x[0])
      s+=float(x[1])
    d[tuple(l)]=round((s/len(l)),2)
    self.grades[sem]=d
  def printDetails(self):
    print('Name:',self.name)
    print('ID:',self.id)
    for i,j in self.grades.items():
      print(f'Courses taken in {i}:')
      for k,l in j.items():
        for m in k:
          print(m)
        print('GPA:',l)
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
#Task 22
class Hotel:
  def __init__(self,hotel_name):
    self.hotel_name=hotel_name
    self.stuff=0
    self.guest=0
    self.stuffs={}
    self.guests={}
  def addStuff(self,name,age=00,phone="000"):
    self.stuff+=1
    self.stuffs[self.stuff]=[name,age,phone]
    print(f'Staff With ID {self.stuff} is added')
  def addGuest(self,name,age=00,phone="000"):
    self.guest+=1
    self.guests[self.guest]=[name,age,phone]
    print(f'Guest With ID {self.guest} is created')
  def getStuffById(self,no):
    x=self.stuffs.get(no)
    return f'Staff ID: {no}\nName: {x[0]}\nAge: {x[1]}\nPhone no.: {x[2]}'
  def getGuestById(self,no):
    y=self.guests.get(no)
    return f'Staff ID: {no}\nName: {y[0]}\nAge: {y[1]}\nPhone no.: {y[2]}'
  def allStaffs(self):
    print('All Staffs:')
    print('Number of Staff:',len(self.stuffs))
    for i,j in self.stuffs.items():
      print(f'Staff ID: {i} Name: {j[0]} Age: {j[1]} Phone no: {j[2]}')
  def allGuest(self):
    print('All Guest:')
    print('Number of Guest:',len(self.guests))
    for i,j in self.guests.items():
      print(f'Guest ID: {i} Name: {j[0]} Age: {j[1]} Phone no.: {j[2]}')
h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest("Carol",35,"123")
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, "431")
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```

```python
#Task 24
class Hospital:
  def __init__(self,name):
    self.name=name
    self.doctors={}
    self.patients={}
  def addDoctor(self,doctor):
    self.doctors[doctor.id]=[doctor.name,doctor.speciality]
  def getDoctorByID(self,id):
    if id in self.doctors:
      return f"Doctor's ID: {id}\nName: {self.doctors[id][0]}\nSpeciality: {self.doctors[id][1]}"
    else:
      return "Doctor not found"
  def allDoctors(self):
    print("All Doctors:")
    print("Number of Doctors:",len(self.doctors))
    print(self.doctors)
  def addPatient(self,patient):
    self.patients[patient.id]=[patient.name,patient.age,patient.phone]
  def getPatientByID(self,id):
    if id in self.patients:
      return f"Patient's ID: {id}\nName: {self.patients[id][0]}\nAge: {self.patients[id][1]}\nPhone no.: {self.patients[id][2]}"
    else:
      return "Patient not found"
  def allPatients(self):
    print("All Patients:")
    print("Number of Patients:",len(self.patients))
    print(self.patients)
class Doctor:
  def __init__(self,id,title,name,speciality):
    self.id=id
    self.title=title
    self.name=name
    self.speciality=speciality
class Patient:
  def __init__(self,id,title,name,age,phone):
    self.id=id
    self.title=title
    self.name=name
    self.age=age
    self.phone=phone
h = Hospital("Evercare")
d1 = Doctor("1d","Doctor", "Samar Kumar", "Neurologist")
h.addDoctor(d1)
print("=================================")
print(h.getDoctorByID("1d"))
print("=================================")
p1 = Patient("1p","Patient", "Kashem Ahmed", 35, 12345)
h.addPatient(p1)
print("=================================")
print(h.getPatientByID("1p"))
print("=================================")
p2 = Patient ("2p","Patient", "Tanina Haque", 26, 33456)
h.addPatient(p2)
print("=================================")
print(h.getPatientByID("2p"))
print("=================================")
h.allDoctors()
h.allPatients()
```

```python
#Task 25
class Vaccine:
  def __init__(self,name,country,interval):
    self.name=name
    self.country=country
    self.interval=interval
class Person:
  def __init__(self,name,age,ptype="General Citizen"):
    self.name=name
    self.age=age
    self.ptype=ptype
    self.vaccines=[]
    self.dose1=False
    self.dose2=False
  def pushVaccine(self,vaccine,dose='1st Dose'):
    self.vaccine_name=vaccine.name
    self.vaccine_interval=vaccine.interval
    self.dose=dose
    if self.age<25 and self.ptype!='Student':
      print(f'Sorry {self.name}, Minimum age for taking vaccines is 25 years now.')
    elif self.age>=25 and self.ptype!='Student' and self.dose=='1st Dose':
      print('1st dose done for',self.name)
      self.dose1=True
      self.vaccines.append(vaccine)
    elif self.ptype=='Student' and self.dose=='1st Dose':
      print('1st dose done for',self.name)
      self.dose1=True
      self.vaccines.append(vaccine)
    elif self.dose=='2nd Dose' and self.vaccines[-1]!=vaccine:
      print(f'Sorry {self.name}, you cant take 2 different vaccines')
    else:
      print('2nd dose done for',self.name)
      self.dose2=True
      self.vaccines.append(vaccine)
  def showDetail(self):
    print(f'Name: {self.name} Age: {self.age} Type: {self.ptype}')
    print('Vaccine name:',self.vaccine_name)
    if self.dose1==True and self.dose2==False:
      print('1st dose: Given')
      print(f'2nd dose: Please come after {self.vaccine_interval} days')
    elif self.dose1==True and self.dose2==True:
      print('1st dose: Given')
      print('2nd dose: Given')
    else:
      print('1st dose: Not Given')
      print('2nd dose: Not Given')
astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
