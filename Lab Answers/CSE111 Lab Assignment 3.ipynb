```python
#Task 2
class Joker:
  def __init__(self,name,power,is_he_psycho):
    self.name=name
    self.power=power
    self.is_he_psycho=is_he_psycho
j1 = Joker('Heath Ledger', 'Mind Game', False)
print(j1.name)
print(j1.power)
print(j1.is_he_psycho)
print("=====================")
j2 = Joker('Joaquin Phoenix', 'Laughing out Loud', True)
print(j2.name)
print(j2.power)
print(j2.is_he_psycho)
print("=====================")
if j1 == j2:
  print('same')
else:
  print('different')
j2.name = 'Heath Ledger'
if j1.name == j2.name:
  print('same')
else:
  print('different')
print(f'''\nThe first if/else block prints out the output as different beacause,
j1 and j2 objects have different memory location''')
print(f'''\nThe second if/else block prints out the output as same because,
we changed the value of the name variable of j2 object. 
So, j1 and j2 objects have same value in their name variable.''')
```

```python
#Task 3
class Pokemon:
  def __init__(self,name1,name2,power1,power2,damage_rate):
    self.pokemon1_name=name1
    self.pokemon2_name=name2
    self.pokemon1_power=power1
    self.pokemon2_power=power2
    self.damage_rate=damage_rate
team_pika = Pokemon('pikachu', 'charmander', 90, 60, 10)
print('=======Team 1=======')
print('Pokemon 1:',team_pika.pokemon1_name,
team_pika.pokemon1_power)
print('Pokemon 2:',team_pika.pokemon2_name,
team_pika.pokemon2_power)
pika_combined_power = (team_pika.pokemon1_power +
team_pika.pokemon2_power) * team_pika.damage_rate
print('Combined Power:', pika_combined_power)
team_bulba = Pokemon('bulbasaur', 'squirtle', 80, 70, 9)
print('=======Team 2=======')
print('Pokemon 1:',team_bulba.pokemon1_name,
team_bulba.pokemon1_power)
print('Pokemon 2:',team_bulba.pokemon2_name,
team_bulba.pokemon2_power)
bulba_combined_power = (team_bulba.pokemon1_power +
team_bulba.pokemon2_power) * team_bulba.damage_rate
print('Combined Power:', bulba_combined_power)
```

```python
#Task 8
class Calculator:
  def __init__(self,num1,op,num2):
    self.num1=num1
    self.op=op
    self.num2=num2
    print('Let’s Calculate!')
    print('Value 1:',self.num1)
    print('Operator:',self.op)
    print('Value 2:',self.num2)
  def add(self):
    print('Result:',self.num1+self.num2)
  def subtract(self):
    print('Result:',self.num1-self.num2)
  def multiply(self):
    print('Result:',self.num1*self.num2)
  def divide(self):
    print('Result:',self.num1/self.num2)
n1=int(input())
op=input()
n2=int(input())
c=Calculator(n1,op,n2)
if op=='+':
  c.add()
if op=='-':
  c.subtract()
if op=='*':
  c.multiply()
if op=='/':
  c.divide()
```

```python
#Task 11
class Shape:
  def __init__(self,name,x,y):
    self.name=name
    self.x=x
    self.y=y
  def area(self):
    if self.name=='Triangle':
      print('Area:',self.x*self.y/2)
    elif self.name=='Square':
      print('Area:',self.x*self.y)
    elif self.name=='Rhombus':
      print('Area:',self.x*self.y/2)
    elif self.name=='Rectangle':
      print('Area:',self.x*self.y)
    else:
      print('Area: Shape unknown')
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#Task 12
class Calculator:
  def __init__(self):
    print('Calculator is ready!')
  def calculate(self, n1, n2, op):
    if op=='+':
      self.val = n1+n2
    elif op=='-':
      self.val = n1-n2
    elif op=='*':
      self.val = n1*n2
    elif op=='/':
      self.val = n1/n2
    self.calculation = f'{n1} {op} {n2} = {self.val}'
    return self.val
  def showCalculation(self):
    print(self.calculation)
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```
